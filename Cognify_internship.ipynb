**Task 2**
Gender Distribution

Gender distribution analysis was performed using the gender column. The frequency of male and female participants was calculated using value_counts(). Bar chart and pie chart visualizations were created using Matplotlib to better understand participant distribution.

```python
df['gender'].value_counts()
```

```python
import matplotlib.pyplot as plt

gender_count = df['gender'].value_counts()

gender_count.plot(
    kind='bar',
    color=['hotpink', 'royalblue'],
    edgecolor='black'
)

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.show()
```

```python
df['gender'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Gender Distribution")
plt.ylabel("")

plt.show()
```

**Task 8** — Investment Duration

```python
df['Duration'].value_counts()
```

```python
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}

df['Duration_Numeric'] = df['Duration'].map(duration_mapping)

df[['Duration', 'Duration_Numeric']].head()
```

```python
average_duration = df['Duration_Numeric'].mean()

print("Average Investment Duration:", average_duration)
```

```python
df['Duration'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title("Investment Duration Distribution")
plt.ylabel("")

plt.show()
```

**Task 10** - Correlation Analysis

```python
numerical_data = df.select_dtypes(include=['int64', 'float64'])

numerical_data.head()
```

```python
correlation_matrix = numerical_data.corr()

correlation_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Analysis Heatmap")

plt.show()
```
